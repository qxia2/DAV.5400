## Scope of Work
### Introduction
MD Squared Property Group is a New York-based property management company that was found in early 2014. The goal of the company is to provide reliable and consistent top-quality management services to the individual or group homeowners, leasees, landlords, tenants, shareholders, and sublets. After a combined three decades in the real estate industry, the company recognized the need for personalized management solutions and have dedicated itself to anticipating and addressing the specific needs of the stakeholders in this industry. 

### Context
In recent years, the company experienced rapid expansion and eager to acquire the business in New York. However, the loss of the building management agreement in most recent couple month shows that the fast-speed expansion might have some negative influence on the quality of services rendered and this might lead to the lost on the company's profit eventually.

### Needs
To make the company profitable by acquiring new business while keeping the service rendered in high quality is the prime goal for the company.  As the person who is communicating with the board of the homeowner, tenants, and so on, the property manager plays an extremely important role in the property management business. Therefore, finding a way to build professionalism among the property managers and improve the satisfaction rate of the stakeholders, especially tenants, become the most important mission for the company.

### Vision
After several company-wide staff meeting, the top management of the company decides to solve the dilemma the company faced above in an analytical way. As a data analyst intern who works for the company, I suggest launching a project to study what are the important factors that may affect the satisfaction rate of the stakeholders. In order to help the property managers to improve professionalism, the company wants to ensure that each property manager understands his or her strength and weakness when rendering service to the customer. Moreover, the company would like the property manager to have improved their work performance by enhancing its strength while strengthening their weakness.

### Outcome
After this project launched the company wants the following outcome delivered:

* A survey that helps to study the tenants’ satisfaction of their property manager.
* A satisfaction analysis that tells the property manager which factor or combination of factors the tenants would like their property manager to have.
* A standard score system (Balanced Scorecard or OCR) used to evaluate the property manager’s work.
* A dashboard that shows whether the property manager improved their service to the tenants by comparing the score of the current month to the previous month.


## Data Dictionary

The data dictionary contains two parts and there are **Name** and **Defination**.

**nu_household**: This column contains the number of the household of the respondents. This column should contain the only integer.

**contact_fre**: The frequency that the tenants contact his or her property manager in the period of one month. This column should contain only "1" to "5". "1" represent always while "5" represent never. One thing to notice is that "1" through "5" are strings.

**ease_contact**: This shows how easy that the tenants could reach out to their property manager. This column should contain only "1" to "5". "1" is strongly satisfied, while "5" is strongly unsatisfied. All elements in this column are strings.

**follow_up**: This shows how the tenants feel satisfied with the follow-up procedures that rendered by their property manager. "1" represents strongly satisfied, while "5" represents strongly unsatisfied. All elements in this column are strings.

**consideration**: This shows how the tenants feel satisfied with the consideration of their property manager when communicating with him or her."1" represents strongly satisfied, while "5" represents strongly unsatisfied. All elements in this column are strings.

**response**: This shows how the tenants feel satisfied with quick responses of their property manager. "1" represents strongly satisfied, while "5" represents strongly unsatisfied. All elements in this column are strings.

**communication**: This shows how the tenants feel about the timeliness, accuracy, and quality of the communication between them and their property manager. "1" represents strongly satisfied, while "5" represents strongly unsatisfied. All elements in this column are strings.

**overall**: This shows how the tenants feel satisfied with the overall service quality of the property management team. "1" represents strongly satisfied, while "5" represents strongly unsatisfied. All elements in this column are strings.

**improvement**: This column shows what aspect that the tenants would like to see the improvement. The column contains availability, responsiveness, problem resolution, communication, and friendliness. All elements in this column are strings.

### Good practice in the conduct and reporting of survey research

https://academic.oup.com/intqhc/article/15/3/261/1856193

* Understand three types of research, and they are descriptive research, analytical studies, and evaluation research.

* The advantages and disadvantages of using survey gather the data.
Advantages
  1. Produces data based on real-world observations
  2. Easier to obtain data based on a representative sample, generalizable to a population
  3. Produce a large amount of data in a short time for a low cost
Disadvantages
  1. The significance of the data can become neglected
  2. The data produced are likely to lack details
  3. Securing a high response rate to a survey can be hard to control 
  
* The aspects should e considered in research question design:
  1. Be knowledgeable about the area you wish to research
  2. Widen the base of your experience, explore related areas, and, communicate
  3. Consider using techniques for enhancing creativity (e.g. Brainstorming)
  4. Avoid the pitfalls of: 
          a. Allowing a decision to decide the questions to be asked 
          b. Posing research questions that cannot be answered
          c. Asking questions that have already been answered satisfactorily

Analysis
