## Part 4: Inference

### 1.  How different country behave differently in success rate

```python
# Slice the series with suceess rate of different countries
Succ_num = (df[df['status']=='successful']['country'])
Total_num = (df['country'])
Succ_rate = Succ_num.value_counts()/Total_num.value_counts()
```

```python
# Turn series into dataframe and then fix the header name
dfs=Succ_rate.to_frame(name='Success rate')
dfs.rename(index=str,columns={"index": "country", "Success rate": "succ_rate"}, inplace=True)
df1 = dfs.reset_index()
df2 = df1.sort_values(by='succ_rate', ascending=False)
```

```python
# make the barplot of Different success Rate of different countries
ax = sns.barplot(x="index", y="succ_rate", data=df2)
ax.set(xlabel="Country", ylabel="Success Rate");
```

Use matplotblib to make a bar plot same as the one above.

```python
# Slice the series with suceess rate of different countries
Succ_num = (df[df['status']=='successful']['country'])
Total_num = (df['country'])
Succ_rate = Succ_num.value_counts()/Total_num.value_counts()
ssr= Succ_rate.sort_values(ascending=False)

# Turn series into dataframe and then fix the header name
dfs=ssr.to_frame(name='Success rate')
dfs.rename(index=str,columns={"index": "country", "Success rate": "succ_rate"}, inplace=True)
dfs.reset_index()

# make the barplot use the matplotlib met
plt.figure(1)

plt.bar(dfs.index,dfs.succ_rate, alpha=0.75,color='g')
plt.xlabel("country")
plt.ylabel("success rate")

plt.show()
```

According to the bar plots above, the country or area with the highest success rate is Hongkong, the country or area with the lowest success rate is Italia. The U.S. is the sixth country in terms of success rate. The success rate is close to 80 percent for Hongkong, less than 40 percent for Italia and close to 65 percent for the U.S.

### 2.Is there any relationship between the goal amount and pledged amount among the succeed cases？

Use the seaborn to draw a scatterplot.

```python
# Slice the series of 'usd_pledged', and 'goal_usd' within successful cases.
df3 = df[df['status']=='successful']
df3[['usd_pledged','goal_usd']]
df3.head(5)

# Draw a scatter plot use seaborn 'sns.lmplot' function.
bx = sns.lmplot(x= 'usd_pledged', y='goal_usd', data=df3)
```

Use the matplotlin to draw a scatterplot.

```python
# Setup the figue.
plt.figure(1)

# Draw a scatter plot with x = 'usd_pledged', and y = 'goal_usd' 
plt.scatter(df3.usd_pledged,df3.goal_usd, alpha=0.50,color='g')
plt.xlabel("Goal")
plt.ylabel("Pledged")
plt.title("Relationship betwenn the Goal and Pledged")

# Draw a trendline on the scatter plot.
z = np.polyfit(df3.usd_pledged, df3.goal_usd, 1)
p = np.poly1d(z)
plt.plot(df3.usd_pledged,p(df3.usd_pledged),"r-")

plt.show()
```

The scatter plots above show a strong positive relationship between the goal amount and the pledged amount. This founding does agree with the correlation statement we made before. After the investigation, we understand that the reason that the correlation table doesn't show the strong relationship between the goal amount and the pledged amount is the effect of the failed cases. Once the failed cases have been eliminated, the more valid relationship between these two variables shown up.

### 3. How Main Category Perform differently on average amount of usd pledged ending in different quarter? 

```python
# Creating a dataframe with the mean of usd pledged of each main category
df5 = df3[['main_category','usd_pledged','end_Q']].round()
pt = df5.pivot_table(index='main_category', columns='end_Q', values='usd_pledged', aggfunc=np.sum)
pt.head()
```

Use the seaborn to draw a heatmap

```python
# creating heatmap of Usd pledged per Main category and End Quarter
f, ax = plt.subplots(figsize = (16, 8))
cmap = sns.cubehelix_palette(start = 1, rot = 3, gamma=1, as_cmap = True)
sns.heatmap(pt, cmap = cmap, linewidths = 1, ax = ax)
ax.set_title('Usd Pledged per Main Category and End Quarter ')
ax.set_xlabel('end_Q')
ax.set_ylabel('main_category')
```

Use the matplotlin to draw a heatmap

```python
# creating heatmap of Usd pledged per Main category and End Quarter
plt.subplots(figsize = (16, 8))
ax = sns.heatmap(pt, linewidth=0.5)
plt.show()
```

The heat-maps above show the main category that pledged most on average is technology. For this specific category, the best ending time should be Q2 and Q4 which those projects usually pledged most on average. On the other hands, the category of design, film & video, games are also performed much better in Q2 to Q4 so that this time period is also great to end a project. However, for other category, this tendency is weak.
