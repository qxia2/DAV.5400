### 1. What proportion of diamonds are between .30 and 1.08 carats? 
First step is to find how many diamonds weight between .30 and 1.08. Then find the total amount of diamonds in data set. The proportion equals the amount of qualified diamonds divided by the total amount of diamond in the data set.

The proportion of diamonds are between .30 and 1.08 carats is 74.68%.

```python
# Use the 'count' function to find the how many diamonds weight between .30 and 1.08
a=df['carat'][(df['carat']<=1.08)&(df['carat']>=0.30)].count()

# Count how many diamond in this dataset in total.
b=df['carat'].count()

# Use a divided b and print out the result
print(f'The proportion of diamonds are between .30 and 1.08 carats is {a/b:.2%}.')
```

### 2. How many of the diamonds have equal x and y dimensions?
Use "count" function to find the amount of the diamonds that have same size in length and width.

There are 17 diamonds have equal x and y dimentsions.

```python
# Use 'count' to find how many diamonds have same dimensions in length and width.
c=df['length'][(df['length']==df['width'])].count()

# Print out the result
print(f'There are {c} diamonds have equal x and y dimentsions.')
```

### 3. How many of the diamonds have a carat less than the mean carat value?

First, find the mean carat value. Then, find the amount of diamonds less than the mean carat value.

There are 31127 diamonds have a carat less than the mean carat value and the mean value is 0.798.

```python
# First things is to calculate the mean value.
d=df['carat'].mean()
d

# Then use 'count' function the find the qualified diamonds.

e=df['carat'][(df['carat']<d)].count()

# print out the result.
print(f'There are {e} diamonds have a carat less than the mean carat value and the mean value is {d:.3}.')
```

### 4. How many diamonds have a Premium cut or better? 
Note that the possible values for the quality of a cut are ranked in ascending order as follows: Fair / Good / Very Good / Premium / Ideal.

First of all, find the diamonds the have a premium or ideal cut. Then, count the row of diamonds that have a premium or ideal cut. 

There are 35,342 diamonds have a Premium cut or better.

```python
# Use 'count' to find how many rows the have Premium or Idealin 'cut' column.
f=df['cut'][(df['cut']=='Premium')|(df['cut']=='Ideal' )].count()
f

# Print out the result.
print(f'There are {f} diamonds have a Premium cut or better.')
```

### 5. Which diamond has the highest price per carat? What is its value?
First, make the price divided bt carat to get the price per carat for each diamond. Then, find the one has the highest value.

The diamond has the highest price per carat is the 27635th, and the value is $18,542 dollars per carat.

```python
# Use index maximum function to identify the diamond that has highest price per carat.
g = (df['price']/df['carat']).idxmax()
g

# Use 'loc' function to show the row of the diamond that has highest price per carat.
df.loc[[g]]

h=df['price'][g]
h
# Print out the result.

print(f'The diamond has the highest price per carat is the {g}th, and the value is ${h} dollars per carat.')
```
