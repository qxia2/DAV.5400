2. Read your .CSV file from your GitHub repository into a Jupyter Notebook, and use your knowledge of combining and reshaping data in Pandas to tidy and transform your data.

```python
# Read .CSV file

df = pd.read_csv('https://raw.githubusercontent.com/qxia2/DAV.5400/master/flight_info.csv')

df
```

First do some data cleaning.

```python
# Add Title for column 2 and change the column head from 'Index Title' to 'Arrive'.

df.rename(columns={'Index Title':'Airlines', ' ':'Status'}, inplace=True)

df
```

Reshape the df into "long" version. Use "Airlines" and "Status" as the keys, use five cities as variables, and use the values of five cities as "value".

```python
# Use "melt" funtion.
long_ver = df.melt(id_vars = ['Airlines','Status'], value_vars = ['Los Angeles', 'Phoenix', 'San Diego', 'San Francisco', 'Seattle'])
long_ver
```

Looks like there are some blanks exist. Therefore I decide to fill the blanks use NaN.

```python
# Fill blanks.
long_ver.replace(' ',np.nan, inplace = True)
long_ver
```

Once the blanks have been filled. I decide to use the "fillna" function to clean "Airlines" column.

```python
# fill the first column.
long_ver['Airlines'].fillna(method='ffill', inplace = True)
long_ver
```

# Change the data type


```python
long_ver.dtypes
```

After changing the data type I found that the "value" column is not the integer it supposes to be. So, I decided to remove the rows that still contain the NaN value change "value" to "Number of Flights", and change the value to an integer.

```python
# Drop the rows still have NaN.
long_ver.dropna(axis='rows', inplace = True)
long_ver
```

```python
# Change the value to integer and add back to long_ver df.
long_ver['Number of Flights'] = long_ver["value"].astype(int)

# Delete the original "value" column.
long_ver.drop(columns=['value'], inplace = True)
long_ver
```

```python
# Now the datatpye for column "Number of Flights" should be integer
long_ver['Number of Flights'].dtypes
```

Next Step is to change the column head to make the table have clear insights. 

```python
# Rename the "variable"Column.
long_ver.rename(columns={'variable':'Cities'}, inplace=True)
long_ver
```
